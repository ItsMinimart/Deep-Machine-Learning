```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as panda 

housing = panda.read_csv("housing.csv")
print('housing shape:', housing.shape)
housing.head()
```

```python
varlist = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']

def binary_map(x):
    return x.map({'yes': 1, 'no': 0})
housing[varlist] = housing[varlist].apply(binary_map)

housing['furnishingstatus'] = housing['furnishingstatus'].replace(['unfurnished'], 0)
housing['furnishingstatus'] = housing['furnishingstatus'].replace(['semi-furnished'], 1)
housing['furnishingstatus'] = housing['furnishingstatus'].replace(['furnished'], 2)

housing.head()
```
