```python
housing = panda.DataFrame(panda.read_csv("housing.csv"))
varlist = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']
def binary_map(x):
    return x.map({'yes': 1, "no": 0})
housing[varlist] = housing[varlist].apply(binary_map)

housing['furnishingstatus'] = housing['furnishingstatus'].replace({'unfurnished':0, 'semi-furnished':1, 'furnished':2})

housing.head()
```
